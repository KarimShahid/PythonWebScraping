```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.fragrancenet.com/fragrances?page=1'
soup = BeautifulSoup(url, 'html.parser')
data = requests.get(soup)
print(data)
```

```python
print(data.text)
```

```python
soup = data.text
htmltags = BeautifulSoup(soup, "lxml")
htmltags
```

```python
lists = htmltags.find_all('div',class_='resultItem heightSync')
lists
```

```python
infos = []
for i in lists:
    name = i.find('span', class_='brand-name').get_text()
    brand = i.find('p', class_='des').get_text().replace('by','').replace('\n','')
    quality= i.find('p', class_='desc').get_text()
    gender = i.find('span', class_='gender-badge').get_text()
# #     print(title)
#     print(price)
#     print(area)
#     print(location)

    info = [name,brand,quality,gender]
    infos.append(info)
print(infos)

#     print(name)
#     print(brand)
#     print(quality)
#     print(gender)
   
```

```python
import csv

f = open('Fragranc_net.csv','w')
x = csv.writer(f)
header = ['Product','Brand','Intensity','Gender']
x.writerow(header)
for i in infos:
    x.writerow(i)
f.close()
```
