```python
from bs4 import BeautifulSoup
import requests
url = "https://www.pararius.com/apartments/amsterdam?ac=1"
soup = BeautifulSoup(url, 'html.parser')
data = requests.get(soup)
print(data)
```

```python
soup = data.text
soup
```

```python
htmltags = BeautifulSoup(soup, "lxml")
htmltags
```

```python
htmltags.title
```

```python
lists = htmltags.find_all('section',class_='listing-search-item')
lists
```

```python
infos = []
for i in lists:
    title = i.find('a', class_='listing-search-item__link listing-search-item__link--title').get_text().replace('\n',' ')
    price = int(i.find('div', class_='listing-search-item__price').get_text().replace('\n','').replace('per month','').replace('€','').replace(',',''))
    location= i.find('div', class_='listing-search-item__location').get_text()
    area = i.find('li', class_='illustrated-features__item illustrated-features__item--surface-area').get_text().replace('m²','')
#     print(title)
#     print(price)
#     print(area)
#     print(location)

    info = [title,location.strip(),price,int(area.strip())]
    infos.append(info)
print(infos)
   
```

```python
import csv

f = open('Housing.csv','w')
x = csv.writer(f)
header = ['Title','Location','Price €','Area m²']
x.writerow(header)
for i in infos:
    x.writerow(i)
f.close()
```
